```python
L=[11,12,13,14]
print("Original List : ",L,"\n")
# (i)
L.append(50)
L.append(60)
print("(i)    --> After adding 50 & 60        : ",L)
# (ii)
L.remove(11)
L.remove(13)
print("(ii)   --> After removing 11 & 13      : ",L)
# (iii)
L.sort()
print("(iii)  --> After sorting in asc order  : ",L)
# (iv)
L.sort(reverse=True)
print("(iv)   --> After sorting in desc order : ",L)
# (v)
if (13 in L):
  print("(v)    --> 13 in L?                    : Present")
else:
  print("(v)    --> 13 in L?                    :  Not Present")  
# (vi)
print("(vi)   --> No of elements              : ",len(L))
# (vii)
print("(vii)  --> Sum of all elements         : ",sum(L))
# (viii)
odd_sum=0
even_sum=0
non_prime_sum=0
for i in L:
  if(i%2!=0):
    odd_sum+=i

  else:
    even_sum+=i 
  for j in range(2,i//2+1):
    if (i%j==0):
      non_prime_sum+=i
    break

   
print("(viii) --> Sum of all odd numbers      : ",odd_sum)
# (ix)
print("(ix)   --> Sum of all even numbers     : ",even_sum)
# (x)
print("(x)    --> Sum of all prime numbers    : ",sum(L)-non_prime_sum) 
# (xi)
L.clear()
print("(xi)   --> After clearing all elements : ",L)
# (xii)
del(L)
try:
  print("(xii)  --> After deleting List L      : ",L)
except:
  print("(xii)  --> After deleting list L       :  List not found.")  
```

```python
import random as r
L=[]
even=[]
odd=[]
prime_nos=[]

def prime(n):
  for i in range(2,n//2+1):
    if (n%i==0):
      return False
  return True

for i in range(100):
  L.append(r.randint(100, 900))
odd_count=0
even_count=0
prime_count=0


for i in L:
  if i%2!=0:
    odd_count+=1
    odd.append(i)
  else:
    even_count+=1
    even.append(i) 
  if prime(i):
    prime_count+=1;
    prime_nos.append(i)    

print("(i)   --> No. of Odd numbers   - %d : " %(odd_count),odd) 
print("(ii)  --> No. of Even numbers  - %d : " %(even_count),even)  
print("(iii) --> No. of Prime numbers - %d : " %(prime_count),prime_nos) 
```

```python
import random as r
import matplotlib.pyplot as plt
L1=[]
L2=[]
L3=[]
L4=[]
L5=[]
list_sizes = [5000, 10000, 15000, 20000, 25000]
times_taken = []

for i in range(5000):
  L1.append(r.randint(-15000,15000))
for i in range(10000):
  L2.append(r.randint(-15000,15000))
for i in range(15000):
  L3.append(r.randint(-15000,15000))
for i in range(20000):
  L4.append(r.randint(-15000,15000))
for i in range(25000):
  L5.append(r.randint(-15000,15000))        

def calculate_time_list():
  start_time = time.time() 
  sorted(L1)
  end_time = time.time()
  time_taken = (end_time - start_time)*1000
  print("Time taken to sort %d elements in a list : %f ms."%(len(L1),time_taken))
  times_taken.append(time_taken)

  start_time = time.time() 
  sorted(L2)
  end_time = time.time()
  time_taken = (end_time - start_time)*1000
  print("Time taken to sort %d elements in a list : %f ms."%(len(L2),time_taken))
  times_taken.append(time_taken)

  start_time = time.time() 
  sorted(L3)
  end_time = time.time()
  time_taken = (end_time - start_time)*1000
  print("Time taken to sort %d elements in a list : %f ms."%(len(L3),time_taken))
  times_taken.append(time_taken)

  start_time = time.time() 
  sorted(L4)
  end_time = time.time()
  time_taken = (end_time - start_time)*1000
  print("Time taken to sort %d elements in a list : %f ms."%(len(L4),time_taken))
  times_taken.append(time_taken)

  start_time = time.time() 
  sorted(L5)
  end_time = time.time()
  time_taken = (end_time - start_time)*1000
  print("Time taken to sort %d elements in a list : %f ms."%(len(L5),time_taken))
  times_taken.append(time_taken)

calculate_time_list()  

plt.plot(list_sizes, times_taken, marker='o')
plt.xlabel('Number of Elements in List')
plt.ylabel('Time Taken (ms)')
plt.title('Time Taken for Sorting Lists')
plt.show()

```

```python
marks_dict = {
    "A": [80, 75, 90, 85, 95],
    "B": [70, 85, 80, 90, 75],
    "C": [80, 85, 80, 75, 70],
    "D": [85, 90, 85, 80, 75],
    "E": [80, 90, 65, 86, 70]
}
max_marks=-1
min_marks=101
for name, marks in marks_dict.items():
  average_marks=sum(marks)/len(marks)
  print(name,average_marks)
  if average_marks>max_marks:
    max_marks=average_marks
    max_avg_student=name
  if average_marks<min_marks:
    min_marks=average_marks 
    min_avg_student=name


print("Student with maximum average marks:", max_avg_student)
print("Student with minimum average marks:", min_avg_student)

```
